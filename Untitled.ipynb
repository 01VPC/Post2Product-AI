```python
# Cell 2: Import libraries
from transformers import pipeline, BlipProcessor, BlipForConditionalGeneration
from PIL import Image
import requests
from io import BytesIO
import pandas as pd
import re
import torch
```

```python
import os
os.environ["TRANSFORMERS_NO_TF"] = "1"
```

```python
# Cell 3: Zero-shot classifier
class ZeroShotClassifier:
    def __init__(self):
        self.classifier = pipeline(
            "zero-shot-classification",
            model="facebook/bart-large-mnli"
        )
    
    def predict(self, text, labels):
        result = self.classifier(text, labels)
        return result['labels'][0], result['scores'][0]
```

```python
# Cell 4: Sentiment analyzer
class SentimentAnalyzer:
    def __init__(self):
        self.analyzer = pipeline("sentiment-analysis")
    
    def analyze(self, text):
        result = self.analyzer(text)[0]
        return result['label'], result['score']
```

```python
# Cell 5: Image caption generator
class ImageCaptioner:
    def __init__(self):
        self.device = "cuda" if torch.cuda.is_available() else "cpu"
        self.processor = BlipProcessor.from_pretrained("Salesforce/blip-image-captioning-base")
        self.model = BlipForConditionalGeneration.from_pretrained(
            "Salesforce/blip-image-captioning-base"
        ).to(self.device)
    
    def caption_image(self, url):
        response = requests.get(url)
        image = Image.open(BytesIO(response.content)).convert("RGB")
        inputs = self.processor(images=image, return_tensors="pt").to(self.device)
        out = self.model.generate(**inputs)
        return self.processor.decode(out[0], skip_special_tokens=True)
```

```python
# Cell 6: Product description generator
class ProductDescriber:
    def __init__(self, groq_key):
        self.client = Groq(api_key=groq_key)
    
    def generate(self, caption, hashtags):
        prompt = f"Create 3-line product description for: {caption} (Hashtags: {hashtags})"
        response = self.client.chat.completions.create(
            messages=[{"role": "user", "content": prompt}],
            model="llama3-8b-8192"
        )
        return re.sub(r"\\n|\"", "", response.choices[0].message.content)
```

```python
# Cell 7: Main processor
class MediaProcessor:
    def __init__(self, groq_key):
        self.classifier = ZeroShotClassifier()
        self.sentiment = SentimentAnalyzer()
        self.captioner = ImageCaptioner()
        self.describer = ProductDescriber(groq_key)
        self.labels = ["electronics", "fashion", "beauty", "sports"]
    
    def process_csv(self, csv_path):
        df = pd.read_csv(csv_path)
        results = []
        
        for _, row in df.iterrows():
            # Classification
            class_label, _ = self.classifier.predict(
                f"{row['caption']} {row['hashtags']}", 
                self.labels
            )
            print(class_label)
            
            # Sentiment
            sentiment, _ = self.sentiment.analyze(row['caption'])
            
            # Image captioning
            caption = self.captioner.caption_image(row['image_url'])
            
            # Product description
            description = self.describer.generate(row['caption'], row['hashtags'])
            
            results.append({
                "original_caption": row['caption'],
                "category": class_label,
                "sentiment": sentiment,
                "image_caption": caption,
                "product_description": description
            })
        
        return pd.DataFrame(results)
```

```python
# Cell 8: Direct image testing
def test_single_image(image_url, caption=""):
    # add you aapi key here
    processor = MediaProcessor(groq_key="")
    
    # Generate caption
    image_caption = processor.captioner.caption_image(image_url)
    
    # Generate description
    description = processor.describer.generate(caption, "")
    
    return {
        "image_url": image_url,
        "generated_caption": image_caption,
        "product_description": description
    }
```
